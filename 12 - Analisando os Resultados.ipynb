## Etapa 12: Analisando os resultados

```python
import pandas as pd 
```

```python
def destaca_maior(s):
    is_max = s == s.max()
    return ['background-color: #99FF99' if v else '' for v in is_max]

def destaca_menor(s):
    is_min = s == s.min()
    return ['background-color: #99FF99' if v else '' for v in is_min]
```

## Modelos treinados com o dataset dos reviews

```python
resultados = {'Modelo': ['SVM', 'CNN', 'RNN'],
              'Tempo Total de Treinamento': [260, 75, 224],
              'Tempo por Epoca': ['NA', 5, 32],
              'Nr de Epocas': ['NA', 15, 7],
              'Acc Treino': [99.25, 97.43, 97.87],
              'Acc Teste': [94.87, 95.35, 94.86]}
```

```python
df = pd.DataFrame(resultados)
df = df.set_index("Modelo")
df.style.apply(destaca_maior, subset=['Acc Teste']).apply(destaca_menor, subset=['Tempo Total de Treinamento'])
```

Neste estudo a Rede Neural Convolucional se mostrou mais performática, tanto em tempo de treinamento quanto em acurácia, apesar de que podemos considerar um empate tecnico neste último quesito.
